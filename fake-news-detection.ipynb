```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake = pd.read_csv("fake-news-data.csv")
data_true = pd.read_csv("true-news-data.csv")
```

```python
data_fake.head()
```

```python
data_true.head()
```

```python
data_fake["class"] = 0
data_true["class"] = 1
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing = data_fake.tail(10)
for i in range(23480, 23470, -1):
    data_fake.drop([i], axis=0, inplace=True)

data_true_manual_testing = data_true.tail(10)
for i in range(21416, 21406, -1):
    data_true.drop([i], axis=0, inplace=True)
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing["class"] = 0
data_true_manual_testing["class"] = 1
```

```python
data_fake_manual_testing.head()
```

```python
data_true_manual_testing.head()
```

```python
data_merge = pd.concat([data_fake, data_true], axis=0)
```

```python
data_merge.head()
```

```python
data_merge.columns
```

```python
data = data_merge.drop(["title", "subject", "date"], axis=1)
```

```python
data.isna().sum()
```

```python
data = data.sample(frac=1)
```

```python
data.head()
```

```python
data.reset_index(inplace=True)
data.drop(["index"], axis=1, inplace=True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub("\[.*?\]", "", text)
    text = re.sub("\\W", " ", text)
    text = re.sub("https?//\S+|www\.\S+", "", text)
    text = re.sub("<.*?>+", "", text)
    text = re.sub("[%s]" % re.escape(string.punctuation), "", text)
    text = re.sub("\n", "", text)
    text = re.sub("\w*\b\w*", "", text)
    return text
```

```python
data["text"] = data["text"].apply(wordopt)
```

```python
X = data["text"]
y = data["class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(X_train)
xv_test = vectorization.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(xv_train, y_train)
```

```python
lr_pred = lr.predict(xv_test)
```

```python
lr.score(xv_test, y_test)
```

```python
print(classification_report(y_test, lr_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(xv_train, y_train)
```

```python
dt_pred = dt.predict(xv_test)
```

```python
dt.score(xv_test, y_test)
```

```python
print(classification_report(y_test, dt_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(xv_train, y_train)
```

```python
gbc_pred = gbc.predict(xv_test)
```

```python
gbc.score(xv_test, y_test)
```

```python
print(classification_report(y_test, gbc_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train, y_train)
```

```python
rfc_pred = rfc.predict(xv_test)
```

```python
rfc.score(xv_test, y_test)
```

```python
print(classification_report(y_test, rfc_pred))
```

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n ==1:
        return "Not A Fake News"

def testing(news):
    testing = {"text": [news]} 
    new_def_test = pd.DataFrame(testing) 
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    lr_pred = lr.predict(new_xv_test)
    dt_pred = dt.predict(new_xv_test)
    gbc_pred = gbc.predict(new_xv_test)
    rfc_pred = rfc.predict(new_xv_test)

    return print("\n\nLR Predication: {} \nDT Predication: {} \nGBC Predication: {} \nRFC Predication: {}".format(
        output_label(lr_pred[0]),
        output_label(dt_pred[0]),
        output_label(gbc_pred[0]), 
        output_label(rfc_pred[0])
    ))
```
